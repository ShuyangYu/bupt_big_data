## 词云

```python
# by暂时不处理，nd采用‘ ’取代
nd_books_borrow_record_data = nd_books_borrow_record_data.fillna(' ')
```

### 直接使用wordcloud进行词云统计

```python
# 直接使用wordcloud进行词云统计
wordcloud_by = WordCloud(font_path='C:/Users/Shuyang Yu/Downloads/simfang.ttf',
                                background_color='black',
#                                 mask=coloring,
                                max_words=200,
                                max_font_size=250,
                                width=1800,
                                height=800,
                                random_state=5603).generate(' '.join(by_books_borrow_record_data['book_name']))
#                                 random_state=5603).generate_from_frequencies(by_books_borrow_record_data['book_name'].count())

stopword_nd = ['第2版', '第3版', '第4版', '第5版', '第6版', 'of', 'and', 'the', 'in', 'to', 'for', '第7版', '修订版']
wordcloud_nd = WordCloud(font_path='C:/Users/Shuyang Yu/Downloads/simfang.ttf',
                                background_color='black',
                                stopwords=stopword_nd,
#                                 mask=coloring,
                                max_words=200,
                                max_font_size=250,
                                width=1800,
                                height=800,
                                random_state=5603).generate(' '.join(nd_books_borrow_record_data['book_name']))
#                                 random_state=5603).generate_from_frequencies(by_books_borrow_record_data['book_name'].count())gridspec.GridSpec(3,2)

f,ax=plt.subplots(2,1,figsize=(24,18))

ax[0].set_title("wordcloud_by ",fontsize=15)
ax[0].imshow(wordcloud_by)
ax[0].axis('off')

ax[1].set_title("wordcloud_nd ",fontsize=15)
ax[1].imshow(wordcloud_nd)
ax[1].axis('off')

plt.show()
```

### 使用jieba+wordclod进行词云统计

```python
# 使用jieba+wordclod进行词云统计
mylist_by = ' '.join(by_books_borrow_record_data['book_name'])
word_list_by = [" ".join(jieba.cut(mylist_by))]
new_text_by = ' '.join(word_list_by)
wordcloud_jieba_by = WordCloud(font_path='C:/Users/Shuyang Yu/Downloads/simfang.ttf',
                                background_color='black',
                                max_words=200,
                                max_font_size=250, 
                                width=1800,
                                height=800,
                                random_state=3000).generate(new_text_by)

mylist_nd = ' '.join(nd_books_borrow_record_data['book_name'])
word_list_nd = [" ".join(jieba.cut(mylist_nd))]
new_text_nd = ' '.join(word_list_nd)
wordcloud_jieba_nd = WordCloud(font_path='C:/Users/Shuyang Yu/Downloads/simfang.ttf',
                                background_color='black',
                                max_words=200,
                                max_font_size=250,
                                width=1800,
                                height=800,
                                random_state=3000).generate(new_text_nd)

f,ax=plt.subplots(2, 1,figsize=(24,18))

ax[0].set_title("wordcloud__jieba_by ",fontsize=15)
ax[0].imshow(wordcloud_jieba_by)
ax[0].axis('off')

ax[1].set_title("wordcloud_jieba_nd ",fontsize=15)
ax[1].imshow(wordcloud_jieba_nd)
ax[1].axis('off')

plt.show()
```

## time feature

```python
by_books_borrow_record_data_timesort = by_books_borrow_record_data['lend_time'].value_counts().reset_index().sort_values(by='index')
by_books_borrow_record_data_timesort = by_books_borrow_record_data_timesort[by_books_borrow_record_data_timesort['index'] >= '2018-03-04']
```

```python
plt.figure(figsize=(15,10))
ax = sns.barplot(y=by_books_borrow_record_data_timesort['index'], x=by_books_borrow_record_data_timesort['lend_time']) 
ax.set(ylabel='date', xlabel='number', title='lend times')
```

```python
nd_books_borrow_record_data = pd.concat((nd_books_borrow_record_data, nd_books_borrow_record_data['lend_time'].str.split(' ',expand=True)), axis=1)
nd_books_borrow_record_data.columns = ['reader_id', 'book_name', 'time', 'lend_deadline', 'lend_time', 'specific_time']
```

```python
nd_books_borrow_record_data_timesort = nd_books_borrow_record_data['lend_time'].value_counts().reset_index().sort_values(by='index')
nd_books_borrow_record_data_timesort = nd_books_borrow_record_data_timesort[nd_books_borrow_record_data_timesort['index'] >= '2017-08-03']
```
